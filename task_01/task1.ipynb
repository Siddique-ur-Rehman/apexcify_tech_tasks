```python

import pandas as pd
```

```python
data=pd.read_csv('/kaggle/input/student-datas/Student.csv')
df=data.copy()
df.head()
```

```python
df=df.drop('Result',axis=1)
df['average_score']=df[['Maths','Physics','Chemistry']].mean(axis=1)
df.head()
```

```python
df.shape
```

```python
print("Maximum Average Marks:",df['average_score'].max())
print("Minimum Average Marks:",df['average_score'].min())
```

```python

idx = df['average_score'].idxmax()


top_student = df.loc[idx]

print("Student with the highest average:")
print(top_student)  

print(f" Average: {top_student['average_score']:.2f}") 
```
